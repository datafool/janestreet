```python
plt.figure(figsize=(20,5))
plt.plot(train_df.ts_id, train_df.resp.cumsum())
plt.xlabel("time")
plt.ylabel("Cummulative Return");
plt.title("Overall Cummulative Return");
```

```python
#Plotting cummulative return for different time horizon
plt.figure(figsize=(20,5))
plt.plot(train_df.ts_id, train_df.resp.cumsum(), label="resp")
plt.plot(train_df.ts_id, train_df.resp_1.cumsum(), label="resp_1")
plt.plot(train_df.ts_id, train_df.resp_2.cumsum(), label="resp_2")
plt.plot(train_df.ts_id, train_df.resp_3.cumsum(), label="resp_3")
plt.plot(train_df.ts_id, train_df.resp_4.cumsum(), label="resp_4")
plt.xlabel("time")
plt.ylabel("Cummulative Return");
plt.title("Overall Cummulative Return");
plt.legend();
```

- Cummulative return of resp_1 < resp_2 < resp_3 < resp_4
